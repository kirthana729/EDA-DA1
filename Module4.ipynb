```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data_path = '/content/elem94_95.csv'
df = pd.read_csv(data_path)

# --- Module 4: Data Summarization and Visualization ---

# 1. Statistical Summary Measures
print("\n--- Statistical Summary Measures ---")
print(df.describe())

# 2. Data Elaboration: Checking unique values, distributions, and anomalies
print("\n--- Data Elaboration ---")
for column in df.select_dtypes(include=['int64', 'float64']).columns:
    print(f"\nColumn: {column}")
    print("Unique Values:", df[column].nunique())
    print("Max Value:", df[column].max())
    print("Min Value:", df[column].min())

# 3. 1-D Statistical Data Analysis
print("\n--- 1-D Statistical Data Analysis ---")
if 'enrol' in df.columns:
    print(df['enrol'].describe())
    plt.figure(figsize=(10, 6))
    sns.histplot(df['enrol'], kde=True, bins=20, color='blue')
    plt.title('1-D Analysis: Distribution of Enrollment')
    plt.xlabel('Enrollment')
    plt.ylabel('Frequency')
    plt.show()

# 4. 2-D Statistical Data Analysis
print("\n--- 2-D Statistical Data Analysis ---")
if {'enrol', 'math4'}.issubset(df.columns):
    plt.figure(figsize=(10, 6))
    sns.scatterplot(x='enrol', y='math4', data=df, color='green')
    plt.title('2-D Analysis: Enrollment vs Math Scores')
    plt.xlabel('Enrollment')
    plt.ylabel('Math Scores')
    plt.show()

# 5. Contingency Tables
print("\n--- Contingency Tables ---")
if {'lunch_bin', 'math4'}.issubset(df.columns):
    contingency_table = pd.crosstab(df['lunch_bin'], pd.cut(df['math4'], bins=4))
    print(contingency_table)

# 6. n-D Statistical Data Analysis
print("\n--- n-D Statistical Data Analysis ---")
if {'enrol', 'math4', 'story4'}.issubset(df.columns):
    sns.pairplot(df[['enrol', 'math4', 'story4']])
    plt.suptitle('n-D Statistical Analysis', y=1.02)
    plt.show()

# 7. Visualization: Scatter Plots, Dot Charts, Bar Plots
# Scatter Plot
if {'math4', 'story4'}.issubset(df.columns):
    plt.figure(figsize=(10, 6))
    sns.scatterplot(x='math4', y='story4', data=df, color='purple')
    plt.title('Scatter Plot: Math Scores vs Story Scores')
    plt.xlabel('Math Scores')
    plt.ylabel('Story Scores')
    plt.show()

# Dot Chart
if 'math4' in df.columns:
    plt.figure(figsize=(10, 6))
    sns.stripplot(x='math4', data=df, color='orange', jitter=True)
    plt.title('Dot Chart: Math Scores')
    plt.xlabel('Math Scores')
    plt.show()

# Bar Plot
if {'lunch_bin', 'math4'}.issubset(df.columns):
    plt.figure(figsize=(10, 6))
    sns.barplot(x='lunch_bin', y='math4', data=df, ci=None, palette='viridis')
    plt.title('Bar Plot: Math Scores by Lunch Categories')
    plt.xlabel('Lunch Categories')
    plt.ylabel('Math Scores')
    plt.show()
```
